### Load the CSV metadata

```python
import numpy as np
import datasets
import matplotlib.pyplot as plt
from IPython.display import display, Markdown
```

## Load Dataset

```python
dataset = datasets.load_dataset("siavava/ai-tech-articles")
df = dataset["train"].to_pandas()
df.head(5)
```

### Print Counts for Each Year


```python
years = df["year"].unique()
years_dfs = {year: df[df["year"] == year] for year in years}

# Number of articles per year
for year in sorted(years, reverse=True):
    display(Markdown(f"## {year} has {len(years_dfs[year])} articles"))
    display(years_dfs[year].head(3))
```

```python
data = np.array(df)
years = data[:, 1]

# get unique years
unique_years = [ int(y) for y in np.unique(years) ]

# get counts
counts = np.array([np.sum(years == year) for year in unique_years])

plt.bar(unique_years, counts, label="Total")
plt.xticks(unique_years)
# change plot size
plt.rcParams['figure.figsize'] = [30, 20]

# show counts on top of bars
for i, v in enumerate(counts):
    plt.text(unique_years[i] - 0.25, v + 0.5, str(v))
plt.legend()
plt.grid()
plt.show()
```
